## Setup

```python
from vanna.ollama import Ollama
from vanna.chromadb import ChromaDB_VectorStore
```

```python
class MyVanna(ChromaDB_VectorStore, Ollama):
    def __init__(self, config=None):
        ChromaDB_VectorStore.__init__(self, config=config)
        Ollama.__init__(self, config=config)

vn = MyVanna(config={'model': 'qwen2.5-coder', 'allow_llm_to_see_data':True})
```

## Training
You only need to train once. Do not train again unless you want to add more training data.

```python

df_ddl = vn.run_sql("SELECT type, sql FROM sqlite_master WHERE sql is not null")

for ddl in df_ddl['sql'].to_list():
  vn.train(ddl=ddl)
```

```python

# The following are methods for adding training data. Make sure you modify the examples to match your database.
schema = r'\nCREATE TABLE materials (\n\tlevel_0 INTEGER, \n\t"index" INTEGER, \n\t"Unnamed: 0" INTEGER, \n\t"BOMcat" TEXT, \n\t"BOM" INTEGER, \n\t"Unnamed: 2" TEXT, \n\t"Material" TEXT, \n\t"Material_Description" TEXT, \n\t"Plant" INTEGER, \n\t"BOM_Usage" INTEGER, \n\t"Alternative" INTEGER, \n\t"Creation_Date" TEXT, \n\t"User" TEXT, \n\t"Component" TEXT, \n\t"Component_Description" TEXT, \n\t"Class" TEXT, \n\t"ICt" TEXT, \n\t"Item" TEXT, \n\t"Quantity" REAL, \n\t"Un" TEXT, \n\t"C.scrap" REAL, \n\t"Prod." TEXT, \n\t"Costing" TEXT\n)\n\n/*\n5 rows from materials table:\nlevel_0\tindex\tUnnamed: 0\tBOMcat\tBOM\tUnnamed: 2\tMaterial\tMaterial_Description\tPlant\tBOM_Usage\tAlternative\tCreation_Date\tUser\tComponent\tComponent_Description\tClass\tICt\tItem\tQuantity\tUn\tC.scrap\tProd.\tCosting\n0\t0\t0\tM\t289476\tStandard\tOFSO1SLYPL\tL1 end chair lobby seats\t1100\t1\t1\t2021-04-22 00:00:00\tPS-DESIGN5\tPKVEB40900200313\t  \tNone\tL\t140\t0.12\tKG\t0.0\tX\tX\n1\t1\t1\tM\t289476\tStandard\tOFSO1SLYPL\tL1 end chair lobby seats\t1100\t1\t1\t2021-04-22 00:00:00\tPS-DESIGN5\tPKPKRW40907101511\t  3.8*66  \tNone\tL\t150\t1.5\tM\t0.0\tX\tX\n2\t2\t2\tM\t289476\tStandard\tOFSO1SLYPL\tL1 end chair lobby seats\t1100\t1\t1\t2021-04-22 00:00:00\tPS-DESIGN5\tSWSWA40500080631\t  8*70\tNone\tL\t160\t2.0\tEA\t0.0\tX\tX\n3\t3\t3\tM\t289476\tStandard\tOFSO1SLYPL\tL1 end chair lobby seats\t1100\t1\t1\t2021-04-22 00:00:00\tPS-DESIGN5\tUAUAW40500030641\t \tNone\tL\t170\t0.03\tKG\t0.0\tX\tX\n4\t4\t4\tM\t289476\tStandard\tOFSO1SLYPL\tL1 end chair lobby seats\t1100\t1\t1\t2021-04-22 00:00:00\tPS-DESIGN5\tMM.AN0294\t  \tNone\tL\t180\t4.0\tEA\t0.0\tX\tX\n*/'
# DDL statements are powerful because they specify table names, colume names, types, and potentially relationships
vn.train(ddl=schema)

# Sometimes you may want to add documentation about your business terminology or definitions.
vn.train(documentation="Our business defines Material as the default order code and Component as default component code")

# You can also add SQL queries to your training data. This is useful if you have some queries already laying around. You can just copy and paste those from your editor to begin generating new SQL.
vn.train(sql="SELECT m.\"Material\" FROM materials m WHERE m.\"Material\" = 'OFSO3SLYPL'")
```

## Asking the AI
Whenever you ask a new question, it will find the 10 most relevant pieces of training data and use it as part of the LLM prompt to generate the SQL.

```python
vn.ask(question="What are the components, descriptions, and quantities in the product with order code: OFDKWDSQLLVN2618FM?")
```
